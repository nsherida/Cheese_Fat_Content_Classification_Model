## Introduction

### Questions of Interest

This analysis will seek to create a model that can predict the fat level of a cheese based on a number of features, including the cheese properties, how it was manufactured, and where it was manufactured. 

Our application will be to support dieticians attempting to categorize cheeses that may not have fat levels explicitly stated on the packaging. More specifically, **our model objective will be to correctly identify as many higher fat cheeses as possible** so that they can be avoided in nutrition plans that are being developed.

### Dataset description


The dataset is composed of 1 table: `cheese_data.csv`, which is from the Government of Canada describing a variety of different Canadian cheeses. It contains a number of columns including:

* `CheeseId` : Unique identifier for each cheese
    
* `ManufacturerProvCode` : Province that each cheese is manufactured in

* `ManufacturingTypeEn` : Manufacturing type for each cheese

* `MoisturePercent` : Percent moisture for each cheese

* `FlavourEn` : A description of the flavour of each cheese

* `CharacteristicsEn` : Miscellaneous characteristics for each cheese

* `Organic` : Binary classification, yes or no (1 or 0) for each cheese

* `CategoryTypeEn` : Category description for each cheese (e.g. soft vs firm)

* `MilkTypeEn` : The type of milk used to produce each cheese (e.g. cow vs goat)

* `MilkTreatmentTypeEn` : The treatment level of each milk (e.g. raw vs pasteurized)

* `RindTypeEn` : The cheese rind type (e.g. washed rind)

* `CheeseName` : The name of each cheese (e.g. cheddar)

* `FatLevel` : Binary class (higher fat or lower fat) for each cheese

The data source can be found [here](https://open.canada.ca/data/en/dataset/3c16cd48-3ac3-453f-8260-6f745181c83b).



### Classification or Regression

The target column for our model is `FatLevel`, which is a binary column with 'higher fat' and 'lower fat' values. As such, this is a classification problem. Based on our application, we will define 'higher fat' as our positive label. 

### Primary Evaluation Metric

As our objective is to identify higher fat cheeses to avoid in nutrition plans, we are looking to maximize the identification of higher fat cheeses. Therefore, our **primary evaluation metric** will be **recall**.

## Methods & Results

With our exploratory analysis complete, we can move into the modeling phase. First, we will split our train and test dfs into X and y, while also dropping the columns discussed above that will not be used for our modelling.

```python
#Split data into X and y, dropping columns discussed previously
X_train = train_df.drop(columns = ['CheeseId', 'FlavourEn', 'CharacteristicsEn', 'CheeseName', 'RindTypeEn', 'FatLevel'])
y_train = train_df['FatLevel']
X_test = test_df.drop(columns = ['CheeseId', 'FlavourEn', 'CharacteristicsEn', 'CheeseName', 'RindTypeEn','FatLevel'])
y_test = test_df['FatLevel']
X_train.head()
```

#### Dummy Model
Next, we can create a dummy classifier to use as a baseline for our model performance. We will score it both on train and test data for comparison later.

```python
#Create dummy classifier, fit on train data, then score accuracy and recall for train and test data
dummy = DummyClassifier(strategy = "most_frequent")
dummy.fit(X_train, y_train)
dummy.score(X_train, y_train)
dummy_output = dummy.predict(X_train)
dummy_output_test = dummy.predict(X_test)
recall_score(y_train, dummy_output, pos_label = 'higher fat')
print('Dummy accuracy:', dummy.score(X_train, y_train).round(2))
print('Dummy recall:', recall_score(y_train, dummy_output, pos_label = 'higher fat').round(2))
print('Dummy test accuracy:', dummy.score(X_test, y_test).round(2))
print('Dummy test recall: ', recall_score(y_test, dummy_output_test, pos_label = 'higher fat').round(2))
```

Our dummy classifier accuracy seems reasonable, as the target was approximately 2/3 lower fat, 1/3 higher fat. The recall of 0 is expected as more than 50% of the dataset is lower fat, so the most frequent dummy model never predicts higher fat.

#### Building Pipeline
Before building our model, we will separate our features into numeric, binary, and categorical.

```python
numeric_features = ["MoisturePercent"]

binary_features = ['Organic']

categorical_features = ['ManufacturerProvCode',
                        'ManufacturingTypeEn', 
                        'CategoryTypeEn', 
                        'MilkTypeEn', 
                        'MilkTreatmentTypeEn']
```

Each of our feature types will require different pre-processing.
- Numeric features: to fill in any null values, we will use simple imputer with strategy = mean. We will then scale the numeric features using standard scaler to ensure the results are not influenced by numeric features with very large or very small magnitudes.
- Categorical features: to fill in any null values, we will use simple imputer with strategy = most frequent. We will then use one hot encoder to convert the string column values into a series of 0-1 columns to ensure all values are scaled equally. 
- Binary features: to fill in any null values, we will use simple imputer with strategy = most frequent. We will then use one hot encoder with drop = if_binary to improve performance (using only one of the mirrored 0-1 columns). Note that since our `Organic` feature is already in 0-1 format we likely don't explicitly need to use one hot encoder, but we will use it for consistency.

Next we will create our transformers for all 3 types of data.

```python
numeric_transformer = make_pipeline(SimpleImputer(strategy = "mean"), 
                                    StandardScaler())

categorical_transformer = make_pipeline(SimpleImputer(strategy = "most_frequent"),
                                        OneHotEncoder(handle_unknown = "ignore"))

binary_transformer = make_pipeline(SimpleImputer(strategy = "most_frequent"),
                                   OneHotEncoder(drop = "if_binary", dtype = int))
```

Combine our column transformers for use in our main pipeline.

```python
preprocessor = make_column_transformer((numeric_transformer, numeric_features),
                                       (categorical_transformer, categorical_features),
                                       (binary_transformer, binary_features))
```

#### Modelling
Now we can build our main pipeline and model. We will use 3 different models and compare their scores to identify the best performing model.

We have selected 3 models: RBF SVM, KNN, and Random Forest. KNN is used as the most simple model with relatively good interpretability. SVM is a more complicated model but still using the similarity/distance concept. Random Forest is more similar to the decision tree type of classifcation rather than using distance. Overall, trying all 3 should give us a relatively wide selection to see which performs best on our dataset.

As discussed in the introduction, our primary scoring metric will be recall. We will create a custom recall scorer for use in our randomized search cv. 

Note: For all 3 models, we have selected randomizedsearchcv for hyperparameter optimization simply for the improvement in execution speed. Gridsearchcv could potentially improve the model performance/scoring, however the server resources were limited. 

#### 1. RBF SVM Model
Build SVC pipeline.

```python
svc_pipe = make_pipeline(preprocessor, SVC(class_weight = "balanced", random_state = 1))
```

Set up hyperparameter optimization for C and gamma, using randomized search to reduce execution time. 

```python
#Define parameters and ranges, create custom scorer for recall, and run hyperparameter optimization
params = {"svc__C":list(np.linspace(0.1, 100, 1000)),
          "svc__gamma": list(np.linspace(0.001, 1, 1000))}

custom_recall = make_scorer(recall_score, pos_label = 'higher fat')
random_svc = RandomizedSearchCV(svc_pipe, params, cv=5, return_train_score = True, verbose = 0, n_jobs = -1, n_iter = 50, scoring = custom_recall, random_state = 1)
random_svc.fit(X_train, y_train)
print("Optimization Complete")
```

Output optimal parameters and scores.

```python
optimal_parameters_svc = random_svc.best_params_
optimal_score_svc = random_svc.best_score_
print('Optimal parameters:', optimal_parameters_svc, "\nOptimal Recall Score:", optimal_score_svc)
```

#### 2. Nearest Neighbours Model
We will now follow the same procedure, but with a kNN model instead.

```python
knn_pipe = make_pipeline(preprocessor, KNeighborsClassifier(weights = "uniform"))
```

```python
#Define parameters and ranges, create custom scorer for recall, and run hyperparameter optimization
params = {"kneighborsclassifier__n_neighbors":list(range(1,100))}

custom_recall = make_scorer(recall_score, pos_label = 'higher fat')
random_knn = RandomizedSearchCV(knn_pipe, params, cv=5, return_train_score = True, verbose = 0, n_jobs = -1, n_iter = 50, scoring = custom_recall, random_state = 2)
random_knn.fit(X_train, y_train)
print("Optimization Complete")
```

```python
optimal_parameters_knn = random_knn.best_params_
optimal_score_knn = random_knn.best_score_
print('Optimal parameters:', optimal_parameters_knn, "\nOptimal Recall Score:", optimal_score_knn)
```

#### 3. Random Forest Model
We will now follow the same procedure, but with a random forest model instead.

```python
rf_pipe = make_pipeline(preprocessor, RandomForestClassifier(class_weight = "balanced"))
```

```python
#Define parameters and ranges, create custom scorer for recall, and run hyperparameter optimization
params = {"randomforestclassifier__n_estimators":list(range(1,500)),
          "randomforestclassifier__max_features":list(range(1,7))}

custom_recall = make_scorer(recall_score, pos_label = 'higher fat')
random_rf = RandomizedSearchCV(rf_pipe, params, cv=5, return_train_score = True, verbose = 0, n_jobs = -1, n_iter = 50, scoring = custom_recall, random_state = 3)
random_rf.fit(X_train, y_train)
print('Optimization Complete')
```

```python
optimal_parameters_rf = random_rf.best_params_
optimal_score_rf = random_rf.best_score_
print('Optimal parameters:', optimal_parameters_rf, "\nOptimal Recall Score:", optimal_score_rf)
```

#### Model Selection
Based on the results of the three modes, kNN provides the best recall score of **0.85** using **nearest neighbours = 2** for a hyperparameter. We can take a look at the classification reports (using train data) for all 3 to confirm our understanding.

```python
print('SVC Train Report:')
print(classification_report(y_train, random_svc.predict(X_train)))
print('KNN Train Report:')
print(classification_report(y_train, random_knn.predict(X_train)))
print('Random Forest Train Report: ')
print(classification_report(y_train, random_rf.predict(X_train)))
```

Based on our training classification reports, it is true that the KNN model provides the highest recall score for our positive label (higher fat) of all 3 models. However, it is also apparent that the KNN model's overall performance is worse than the other two models in terms of accuracy, f1 score, and precision. 

Therefore we will use our KNN model with n_neighbors = 2 as our final optimal model due to its highest recall score. However, if our application/model goal was different, we would likely prefer the random forest model. 

#### Evaluate Model on Test Data
We will now score our KNN model on the test data and view the classification report and the confusion matrix.

```python
#Predict y using our optimal model and X test data. Score accuracy and recall.
pred_knn = random_knn.predict(X_test)
print('Accuracy: ', accuracy_score(y_test, pred_knn))
print('Recall: ', recall_score(y_test, pred_knn, pos_label = 'higher fat'))
```

Our recall score is lower than on our test data, but likely still high enough for our model to be useful and informative.

```python
print('Final Model Classification Report:')
print(classification_report(y_test, random_knn.predict(X_test)))
```

```python
print('Figure 3: Final Model Confusion Matrix')
plot_confusion_matrix(random_knn, X_test, y_test, cmap = "Blues")
```

As expected, our model is maximizing recall score, which leads to a small number of **false negatives** (only **12** predicted lower fat when the true label is higher fat). However, we do have a sacrifice in the form of a higher number of **false positives** (**34** predicted as higher fat when the true label is lower fat).
